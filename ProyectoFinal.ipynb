```python
def menu():

  print("1.- Alta de empleado") 
  print("2.- Calcular IMC de empleados")
  print("3.- Cambia información del empleado")
  print("4.- Consulta un empleado")
  print("5.- Reporte de todos los empleados")
  print("6.- Salir")
  num = int(input("Selecciona un número del menú: "))

  return num

empleados = []

def grabar_archivo(empleados):
  with open("proyectoF.txt", "w") as archivo:
    for empleado in empleados:
      empleado_nuevo = []
      for dato in empleado:
        dato_nuevo = str(dato)
        empleado_nuevo.append(dato_nuevo)
      texto = ",".join(empleado_nuevo)
      #print(texto)
      archivo.write(texto + "\n")

def leer_archivo(empleados):
  with open("proyectoF.txt", "r") as archivo: 
    texto = archivo.readlines()
    lista = []
    for elemento in texto:
      lista.append(elemento.split(","))
    for elemento in lista:
      empleados.append(elemento)
    for i in range(len(empleados)):
      empleados[i][-1] = empleados[i][-1].replace("\n", "")
      
  return empleados



leer_archivo(empleados)





def alta_empleado(empleados):


  id = len(empleados)+1
  nombre = input("Teclea el nombre del empleado: ")
  edad = int(input("Teclea la edad: "))
  peso = float(input("Teclea el peso (kg): "))
  estatura = float(input("Teclea la estatura: "))
  diabetes = int(input("Diabetes 1-Si 0- No: " ))
  hipertencion = int(input("Hipertención 1-Si 0- No: " ))
  corazon = int(input("Enfermedad de corazón 1-Si 0- No: " ))
  cancer = int(input("Cancer 1-Si 0- No: " ))
  tabaquismo = int(input("Tabaquismo 1-Si 0- No: " ))
  vacuna_covid_19 = int(input("Vacuna covid_19 1-Si 0- No: " ))
  departamento = int(input("Número de departamento: " ))
  total = str(id)+","+ nombre +","+ str(edad)+","+ str(peso)+","+ str(estatura)+","+ str(diabetes)+","+ str(hipertencion)+","+ str(corazon)+","+ str(cancer)+","+ str(tabaquismo)+","+ str(vacuna_covid_19)+","+ str(departamento)+"."

  datos_empleado = total.split(",")
  print(datos_empleado)

  empleados.append(datos_empleado)


def calcular_imc(empleados):
  

  contador_imc = [0, 0, 0, 0, 0, 0, 0, 0]
  etiqueta = ["Delgadez severa", "Delgadez moderada", "Delgadez aceptable", "Normal", "Pre-obeso", "Obeso tipo I", "Obeso tipo II", "Obeso tipo III"]

  print(f'{"Clasificación":<20}{"IMC":<10}{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Estatura":<10}{"Enfermedades":<77}{"Vacuna Covid19":<12}')


  for empleado in range(len(empleados)):
    id = empleados[empleado][0]
    nombre =  empleados[empleado][1]
    edad = empleados[empleado][2]
    peso = empleados[empleado][3]
    estatura = empleados[empleado][4]
    diabetes = empleados[empleado][5]
    hipertencion = empleados[empleado][6]
    corazon = empleados[empleado][7]
    cancer = empleados[empleado][8]
    tabaquismo = empleados[empleado][9]
    vacuna_covid_19 = empleados[empleado][10]
    imc = (float(peso) / float(estatura)) ** 2 
    empleados[empleado].append(round(imc,2))
    
    if imc < 16:
      clasificacion = etiqueta[0]
      contador_imc[0] += 1
    elif 16 <= imc and imc < 16.99:
      clasificacion = etiqueta[1]
      contador_imc[1] += 1
    elif 17 <= imc and imc < 18.49:
      clasificacion = etiqueta[2]
      contador_imc[2] += 1
    elif 18.50 <= imc and imc < 24.99:
      clasificacion = etiqueta[3]
      contador_imc[3] += 1
    elif 25 <= imc and imc < 29.99:
      clasificacion = etiqueta[4]
      contador_imc[4] += 1
    elif 30 <= imc and imc < 34.99:
      clasificacion = etiqueta[5]
      contador_imc[5] += 1
    elif 35 <= imc and imc < 40:
      clasificacion = etiqueta[6]
      contador_imc[6] += 1
    else:
      clasificacion = etiqueta[7]
      contador_imc[7] += 1

    imc = round(imc, 2)


    print(f'{clasificacion:<20}{imc:<10}{id:<10}{nombre:<15}{peso:<10}{estatura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertencion == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid_19 == "1" else "____________":15}')
  
  print(f'\n{"Clasificación":27}{"Cantidad":10}{"Porcentaje":<10}')
    

  all = []
  for num in range(len(contador_imc)):
    all_empl = []
    all_empl = [contador_imc[num], etiqueta[num], round(contador_imc[num] / len(empleados) * 100)]
    all.append(all_empl)
  
  all.sort(reverse = True)

  print(f'\n{"Clasificación":27}{"Cantidad":10}{"Porcentaje":<10}')

  for dato in all:
    contador, etiqueta_imc, porcentaje = dato
    print(f"{etiqueta_imc:20} {contador:7} {porcentaje:10}%")
  
  print(f'{"Total de empleados:":<10} {len(empleados):<10}')



def actualizar_empleado(empleados):
  dato = input("Dame tu id: ")
  validacion = False
  for empleado in empleados:
    id = empleado[0]
    if id == dato:
      validacion = True
      id_empleado = int(id)
    if validacion == True:
      print("Si existe")
      print(empleados[id_empleado-1])
      break
  tipo_cambio = int(input("¿Qué quieres cambiar?"))
  if tipo_cambio >= 2 or tipo_cambio <= 4:
    dato_afirmacion = input("¿Está seguro de que quiere cambiar y actualizar nuevos los datos del empleado?: ")
    if dato_afirmacion == "si":
        dato_nuevo = float(input("Ingrese el dato nuevo: "))
        empleados[id_empleado -1][tipo_cambio] = dato_nuevo
    else:
      print("no se econtro")
  print(empleados[id_empleado-1])


def consulta_empleado(empleados):
  dato = input("Dame tu id: ")
  validacion = False
  for empleado in empleados:
    id = empleado[0]
    if id == dato:
      validacion = True
      id_empleado = int(id)
  if validacion == True:
    print("Si existe")
    print(empleados[id_empleado-1])
  else:
      print("No existe")    

def crear_reporte_empleado(empleados):
  contador_enfermedades = [0, 0, 0, 0, 0, 0]
  enfermedad = ["Diabetes", "Hipertensión", "Corazón", "Cáncer", "Tabaquismo", "Vacuna Covid"]

  print(f'{"ID":<10}{"Nombre":<15}{"Peso":<10}{"Altura":<10}{"Enfermedades":^77}{"Vacuna Covid":<12}')

  for empleado in range(len(empleados)):
    id = empleados[empleado][0]
    nombre =  empleados[empleado][1]
    edad = empleados[empleado][2]
    peso = empleados[empleado][3]
    estatura = empleados[empleado][4]
    diabetes = empleados[empleado][5]
    hipertencion = empleados[empleado][6]
    corazon = empleados[empleado][7]
    cancer = empleados[empleado][8]
    tabaquismo = empleados[empleado][9]
    vacuna_covid_19 = empleados[empleado][10] 
    imc = (float(peso) / float(estatura)) ** 2
    empleados[empleado].append(round(imc,2))

    
    if diabetes == "1":
      contador_enfermedades[0] += 1
    if hipertencion == "1":
      contador_enfermedades[1] += 1
    if corazon == "1":
      contador_enfermedades[2] += 1
    if cancer == "1":
      contador_enfermedades[3] += 1
    if tabaquismo == "1":
      contador_enfermedades[4] += 1
    if vacuna_covid_19 == "1":
      contador_enfermedades[5] += 1

    imc = round(imc, 2)


    print(f'{id:<10}{nombre:<15}{peso:<10}{estatura:<10}{"Diabetes" if diabetes == "1" else "____________":15}{"Hipertensión" if hipertencion == "1" else "____________":17}{"Corazón" if corazon == "1" else "____________":15}{"Cáncer" if cancer == "1" else "____________":15}{"Tabaquismo" if tabaquismo == "1" else "____________":15}{"Vacunado" if vacuna_covid_19 == "1" else "____________":15}')
  
  print(f'\n{"Enfermedad":27}{"Cantidad":10}{"Porcentaje":<10}')
    

  all = []
  for num in range(len(contador_enfermedades)):
    all_empl = []
    all_empl = [contador_enfermedades[num], enfermedad[num], round(contador_enfermedades[num] / len(empleados) * 100)]
    all.append(all_empl)
  
  all.sort(reverse = True)


  for dato in all:
    contador, etiqueta_imc, porcentaje = dato
    print(f"{etiqueta_imc:20} {contador:7} {porcentaje:10}%")
  
  print(f'{"Total de empleados:":<10} {len(empleados):<10}')

    

    


def main():
  
  num = menu()
  while num != 7:
    if num == 1:
      print("Alta de empleado") 
      alta_empleado(empleados)
    elif num == 2:
      print("Calcular IMC de empleados")
      calcular_imc(empleados)
    elif num == 3:
      print("Cambia información del empleado")
      actualizar_empleado(empleados)
    elif num == 4:
      print("Consulta un empleado")
      consulta_empleado(empleados)
    elif num == 5:
      crear_reporte_empleado(empleados)
      print("Reporte de todos los empleados")
    elif num == 6:
      print("Salir")
      grabar_archivo(empleados)
      break
    else:
      print("Respuesta Incorrecta")
    num = menu()



main()
```
